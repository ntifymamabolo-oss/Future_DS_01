```python
df=pd.read_csv("/content/sample_data/online_retail.csv (1).zip")
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=["CustomerID"])
df
```

```python
df["InvoiceDate"]=pd.to_datetime(df["InvoiceDate"])
df
```

```python
df["revenue"]=df["Quantity"]*df["UnitPrice"]
df["revenue"]
```

```python
df.info()
```

```python
df["Month"]=df["InvoiceDate"].dt.month
df["Month"]
```

```python
df['InvoiceDate']=pd.to_datetime(df['InvoiceDate'])
df_indexed=df.set_index('InvoiceDate')
df['revenue']
```

```python
monthly_revenue=df_indexed['revenue'].resample('ME').sum()
monthly_revenue
```

```python
monthly_revenue = df.groupby('Month')['revenue'].sum().reset_index()
monthly_revenue
```

```python
import matplotlib.pyplot as plt

plt.figure()
monthly_revenue.plot(figsize=(12,6), marker='o',color="red")
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Monthly Revenue Trend')
plt.savefig('monthly_revenue_trend.png')
plt.show()
```

```python
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
top_products.plot(kind='bar')
plt.title('Top 10 Selling Products (by Quantity)')
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.savefig('top_selling_products.png')
plt.show()
```

```python
region_sales = df.groupby("Country")["revenue"].sum().sort_values(ascending=False)

region_sales.plot(kind='bar')
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.title("Revenue by Region")
plt.savefig("revenue_by_region.png")
plt.show()
```
